```python
df['Education'] = df['Education'].astype(str)
```

```python
df.dtypes
```

```python
def Ecleaner(x):
    if x =='1':
        return x.replace('1',"Below College")
    elif x =='2':
        return x.replace('2','College')
    elif x =='3':
        return x.replace('3','Bachelor')
    elif x=='4':
        return x.replace('4','Master')
    elif x=='5':
        return x.replace('5','Doctor')
        
    else:
        return x    
    
    
```

```python


df['Education'] = df['Education'].apply(Ecleaner)
```

```python
df['Education']
```

```python
df
```

```python
df[df['Attrition']=='Yes'].groupby('Education')['Attrition'].count() # Here we find the the no of attrition per education level
```

```python
def Environment(x):
    if x =='1':
        return x.replace('1',"Low")
    elif x =='2':
        return x.replace('2','Medium')
    elif x =='3':
        return x.replace('3','High')
    elif x=='4':
        return x.replace('4','Very high')
        
    else:

        return x    
```

```python
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].astype(str).apply(Environment)
```
